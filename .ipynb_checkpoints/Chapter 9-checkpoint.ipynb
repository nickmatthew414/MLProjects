```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
olivetti = fetch_olivetti_faces()
```

```python
olivetti.keys()
```

```python
X = olivetti['data']
y = olivetti['target']
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
strat_split = StratifiedShuffleSplit(n_splits=1, test_size=40, random_state=42)
train_valid_idx, test_idx = next(strat_split.split(olivetti.data, olivetti.target))
X_train_valid = olivetti.data[train_valid_idx]
y_train_valid = olivetti.target[train_valid_idx]
X_test = olivetti.data[test_idx]
y_test = olivetti.target[test_idx]
```

```python
strat_split = StratifiedShuffleSplit(n_splits=1, test_size=80, random_state=43)
train_idx, valid_idx = next(strat_split.split(X_train_valid, y_train_valid))
X_train = X_train_valid[train_idx]
y_train = y_train_valid[train_idx]
X_valid = X_train_valid[valid_idx]
y_valid = y_train_valid[valid_idx]
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
pca = PCA(n_components=.99)
X_train_reduced = pca.fit_transform(X_train)
X_valid_reduced = pca.transform(X_valid)
X_test_reduced = pca.transform(X_test)

pca.n_components_
```

```python
k_range = range(5,150,5)
k_means_per_k = []
for k in k_range:
    print('k={}'.format(k))
    kmeans = KMeans(n_clusters=k, random_state=42).fit(X_train_reduced)
    k_means_per_k.append(kmeans)
```

```python
from sklearn.metrics import silhouette_score
import numpy as np
```

```python
silhouette_scores = [silhouette_score(X_train_reduced, model.labels_)
                     for model in k_means_per_k]
best_index = np.argmax(silhouette_scores)
best_k = k_range[best_index]
```

```python
best_score = silhouette_scores[best_index]
```

```python
best_score
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.figure(figsize=(8,3))
plt.plot(k_range, silhouette_scores, 'bo-')
plt.xlabel('$k$', fontsize=12)
plt.ylabel('Silhouette Score', fontsize=12)
plt.plot(best_k, best_score, 'ro-')
plt.show()
```

```python
inertias = [model.inertia_ for model in k_means_per_k]
best_inertia = inertias[best_index]

plt.figure(figsize=(8,3.5))
plt.plot(k_range, inertias, 'bo-')
plt.xlabel('$k$', fontsize=12)
plt.ylabel('Inertia', fontsize=12)
plt.plot(best_k, best_inertia, 'ro-')
```

```python
n_clusters = 10
kmeans = KMeans(n_clusters=n_clusters)
```

```python
best_model = k_means_per_k[best_index]
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=150, random_state=42)
rfc.fit(X_train_reduced, y_train)
rfc.score(X_valid_reduced, y_valid)
```

```python
X_train_transform = best_model.transform(X_train_reduced)
X_valid_transform = best_model.transform(X_valid_reduced)
X_test_transform = best_model.transform(X_test_reduced)
```

```python
rfc.fit(X_train_transform, y_train)
rfc.score(X_valid_transform, y_valid)
```

```python
from sklearn.pipeline import Pipeline

pipeline_scores = []
for n_clusters in k_range:
    pipeline = Pipeline([('kmeans', KMeans(n_clusters=n_clusters, random_state=n_clusters)),
                         ('rfc', RandomForestClassifier(n_estimators=150, random_state=42))])
    pipeline.fit(X_train_reduced, y_train)
    score = pipeline.score(X_valid_reduced, y_valid)
    pipeline_scores.append(score)
print(pipeline_scores)
```

```python
def plot_performance(scores, k_range):
    plt.xlabel('Number of Clusters', fontsize=12)
    plt.ylabel('Scores', fontsize=12)
    plt.plot(k_range, scores, 'ro-')
    plt.show()

plot_performance(pipeline_scores, k_range)
```

```python
rfc.fit(X_train_extended, y_train)
rfc.score(X_valid_extended, y_valid)
```

```python
from sklearn.mixture import GaussianMixture
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=.99)
X_train_pca = pca.fit_transform(X_train)
X_valid_pca = pca.transform(X_valid)
X_test_pca = pca.transform(X_test)
```

```python
gm = GaussianMixture(n_components=3, n_init=10)
gm.fit(X_train_pca, y_train)
```

```python
gm.weights_
```

```python
gm.score(X_valid_pca, y_valid)
```

```python
def plot_faces(faces, labels, n_cols=5):
    n_rows = (len(faces) - 1) // n_cols + 1
    plt.figure(figsize=(n_cols, n_rows * 1.1))
    for index, (face, label) in enumerate(zip(faces, labels)):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(face.reshape(64, 64), cmap="gray")
        plt.axis("off")
        plt.title(label)
    plt.show()
```

```python
n_gen_faces = 20
gen_faces_reduced, y_gen_faces = gm.sample(n_samples=n_gen_faces)
gen_faces = pca.inverse_transform(gen_faces_reduced)

plot_faces(gen_faces, y_gen_faces)
```

```python
n_rotated = 4
rotated = np.transpose(X_train[:n_rotated].reshape(-1, 64, 64), axes=[0, 2, 1])
rotated = rotated.reshape(-1, 64*64)
y_rotated = y_train[:n_rotated]

n_flipped = 3
flipped = X_train[:n_flipped].reshape(-1, 64, 64)[:, ::-1]
flipped = flipped.reshape(-1, 64*64)
y_flipped = y_train[:n_flipped]

n_darkened = 3
darkened = X_train[:n_darkened].copy()
darkened[:, 1:-1] *= 0.3
darkened = darkened.reshape(-1, 64*64)
y_darkened = y_train[:n_darkened]

X_bad_faces = np.r_[rotated, flipped, darkened]
y_bad = np.concatenate([y_rotated, y_flipped, y_darkened])

plot_faces(X_bad_faces, y_bad)
```
